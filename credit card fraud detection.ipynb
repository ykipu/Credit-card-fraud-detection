```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
card_data=pd.read_csv('creditcard.csv')                  # To load the data from csv file
```

```python
card_data.head()
```

```python
card_data.tail()
```

```python
card_data.info()                   # Data set Information
```

```python
card_data.isnull().sum()
```

```python
card_data['Class'].value_counts()            #distribution of right transactions and fraud transactions
```

```python
right=card_data[card_data.Class==0]                #this data base is highly unbalanced to correct this data base we  
fraud=card_data[card_data.Class==1]                # indicate 0=right trans and 1=fraud trans.
```

```python
print(right.shape)
print(fraud.shape)
```

```python
right.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the both transactions
card_data.groupby('Class').mean()
```

```python
right_sample = right.sample(n=492)
```

```python
new_dataset = pd.concat([right_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.count
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model=LogisticRegression()
```

```python
# trainning the locistic Regression model with tarining card
model.fit(x_train,y_train)
```

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy =accuracy_score(x_train_prediction,y_train)
```

```python
print("Accuracy on Training Data :", training_data_accuracy)
```

```python
# accuracy on test data
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print("Accuracy on Training Data of x :",test_data_accuracy)
```
